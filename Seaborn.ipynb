```python
import seaborn as sns
import matplotlib as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### Distribution Plots

```python
sns.distplot(tips['total_bill'],kde=False)
# the curve line is kde (kernel density estimation)
# y axis = count
# x axis = total_bill 
```

```python
sns.jointplot(x='total_bill', y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip',data=tips,kind='reg')
```

```python
# pairplot plots pair wise relationship across entire data set
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

### Categorical Plots

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
# counts no of occurrences
sns.countplot(x='sex',data=tips)
```

```python
# shows distribution of categorical data
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# shows distribution of categorical data
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
#you cannot figure out how many point are stacked on each other. so use jitter
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.factorplot(x='day',y='total_bill',data=tips)
```

### Matrix Plots

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tips.corr()
```

```python
tc = tips.corr()
sns.heatmap(tc, annot=True)
```

```python
flights
```

```python
flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(fp,cmap='coolwarm')
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

### Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
h = sns.FacetGrid(data=tips,col='time',row='smoker')
h.map(sns.distplot,'total_bill')
```

### Regression Plots


```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
#separates the plot based on col parameter. in this case Male and Female.
#hence 2 columns
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',hue='sex')
```

```python
#changing aspect ratio of the plot changes the size
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',hue='sex',
          aspect=0.6,size=8)
```
